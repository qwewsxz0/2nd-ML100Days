```python
import pandas as pd
import numpy as np
import random
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt
```

```python
data = {'weekday': ['Sun', 'Sun', 'Mon', 'Mon'],
        'city': ['Austin', 'Dallas', 'Austin', 'Dallas'],
        'visitor': [139, 237, 326, 456]}
```

```python
visitors_1 = pd.DataFrame(data)
print(visitors_1)
```

```python
cities = ['Austin', 'Dallas', 'Austin', 'Dallas']
weekdays = ['Sun', 'Sun', 'Mon', 'Mon']
visitors = [139, 237, 326, 456]

list_labels = ['city', 'weekday', 'visitor']
list_cols = [cities, weekdays, visitors]

zipped = list(zip(list_labels, list_cols))
print(zipped)
```

```python
visitors_2 = pd.DataFrame(dict(zipped))
print(visitors_2)
```

```python
visitors_1.groupby(by="weekday")['visitor'].mean()
```

```python
data = {'國家':[] ,
        '人口':[]}
country = []
population = []
for country_name in range (0,10):
    country.append(country_name)
print(country_name)
print(country)
data['國家'] = country
print(data['國家'] )
for i in range (0,10):
    population.append(random.randrange(100,1000))
print(population)
data['人口'] = population
print(data['人口'])
data = pd.DataFrame(data)
print(data)
```

```python
print(data['人口'])
print(data[['人口']])##變成list
print(data[['人口']][1:2])##可以取出某一項
print(data[['人口']][0:1] == 855)##可以做bool運算，指示還是塞在DataFram裡面，後面就沒有用此方法了
```

```python
max_population = 0
max_population_country = 0
othoer_max_population_country = []
for i in range(0,10):
    if (data.iloc[i][1] > max_population):
        max_population = data.iloc[i][1]
        max_population_country = i
for i in range(0,10):
    if(data.iloc[i][1] == max_population):
        othoer_max_population_country.append(i)
        

print(max_population)
print(max_population_country)
print(othoer_max_population_country)
```

```python
data = []
with open("C:/Users/IRIS168/imagenet_urls_examples.txt", 'r') as f:
    for line in f:
        line = line.replace('\n', '').replace('\t', ',').split(',') # 將每句最後的 /n 取代成空值後，再以逗號斷句
        data.append(line)
print(data)
```
